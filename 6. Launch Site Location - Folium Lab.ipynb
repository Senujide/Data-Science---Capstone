```python
# TASK 3: Calculate the distances between a launch site to its proximities
# ===============================
# Task 3: Calculate Distances
# ===============================

import folium
from math import sin, cos, sqrt, atan2, radians

# --------------------------------------------------------
# 1. Distance calculation function (Haversine Formula)
# --------------------------------------------------------

def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371.0   # Earth radius in kilometers

    # Convert coordinates to radians
    rlat1 = radians(lat1)
    rlon1 = radians(lon1)
    rlat2 = radians(lat2)
    rlon2 = radians(lon2)

    dlat = rlat2 - rlat1
    dlon = rlon2 - rlon1

    a = sin(dlat/2)**2 + cos(rlat1) * cos(rlat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# --------------------------------------------------------
# 2. Select a launch site (example: CCAFS LC-40)
# --------------------------------------------------------

launch_site = launch_sites_df[launch_sites_df['Launch Site'] == 'CCAFS LC-40']

launch_lat = launch_site.iloc[0]['Lat']
launch_lon = launch_site.iloc[0]['Long']

print("Launch site coordinates:", launch_lat, launch_lon)

# --------------------------------------------------------
# 3. Define proximities (examples)
# --------------------------------------------------------

# Nearest City (Cocoa Beach, FL)
city_coordinate = [28.3200, -80.6076]

# Example Highway location near CCAFS
highway_coordinate = [28.5637, -80.5707]

# Example Railroad near CCAFS
railroad_coordinate = [28.5721, -80.5853]

# --------------------------------------------------------
# 4. Compute Distances
# --------------------------------------------------------

distance_city = calculate_distance(launch_lat, launch_lon, city_coordinate[0], city_coordinate[1])
distance_highway = calculate_distance(launch_lat, launch_lon, highway_coordinate[0], highway_coordinate[1])
distance_railroad = calculate_distance(launch_lat, launch_lon, railroad_coordinate[0], railroad_coordinate[1])

print("Distance to City:", distance_city, "km")
print("Distance to Highway:", distance_highway, "km")
print("Distance to Railroad:", distance_railroad, "km")

# --------------------------------------------------------
# 5. Plot all on map
# --------------------------------------------------------

distance_map = folium.Map(location=[launch_lat, launch_lon], zoom_start=10)

# Marker for launch site
folium.Marker(
    [launch_lat, launch_lon],
    popup='Launch Site: CCAFS LC-40',
    icon=folium.Icon(color='blue')
).add_to(distance_map)

# Marker for city
folium.Marker(
    city_coordinate,
    popup=f"Cocoa Beach (Distance: {distance_city:.2f} km)",
    icon=folium.Icon(color='green')
).add_to(distance_map)

# Marker for highway
folium.Marker(
    highway_coordinate,
    popup=f"Highway (Distance: {distance_highway:.2f} km)",
    icon=folium.Icon(color='purple')
).add_to(distance_map)

# Marker for railroad
folium.Marker(
    railroad_coordinate,
    popup=f"Railroad (Distance: {distance_railroad:.2f} km)",
    icon=folium.Icon(color='red')
).add_to(distance_map)

# Draw lines between launch site and proximities
folium.PolyLine([[launch_lat, launch_lon], city_coordinate], color="green").add_to(distance_map)
folium.PolyLine([[launch_lat, launch_lon], highway_coordinate], color="purple").add_to(distance_map)
folium.PolyLine([[launch_lat, launch_lon], railroad_coordinate], color="red").add_to(distance_map)

# Display map
distance_map
```

```python
# ---- Task 3: Add marker for closest city / railway / highway ----

import folium
from folium.features import DivIcon
from folium.plugins import MousePosition
from math import sin, cos, sqrt, atan2, radians

# Distance calculation (already provided)
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # Earth radius in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    return R * c


# ------------------------------------------------------
# STEP 1: Choose Launch Site (example: CCAFS SLC-40)
# ------------------------------------------------------
launch_lat = 28.563197
launch_lon = -80.576820

city_map = folium.Map(location=[launch_lat, launch_lon], zoom_start=10)

# Add launch site marker
folium.Marker(
    [launch_lat, launch_lon],
    popup="Launch Site",
    icon=folium.Icon(color="blue")
).add_to(city_map)

# Add Mouse Position so you can read coordinates from map
formatter = "function(num){return L.Util.formatNum(num, 5);};"
MousePosition(
    position="topright",
    separator=" | ",
    prefix="Coordinates:",
    lat_formatter=formatter,
    lng_formatter=formatter
).add_to(city_map)


# ------------------------------------------------------
# STEP 2: After selecting city/railway/highway point,
# UPDATE THESE VALUES
# ------------------------------------------------------

# Example: Nearest city coordinate (replace with your chosen point)
closest_point_lat = 28.39220
closest_point_lon = -80.60770

# Calculate distance
distance_to_point = calculate_distance(
    launch_lat, launch_lon,
    closest_point_lat, closest_point_lon
)

# Add marker to that closest point
folium.Marker(
    [closest_point_lat, closest_point_lon],
    popup="Closest City / Railway / Highway",
    icon=folium.Icon(color="red")
).add_to(city_map)

# ------------------------------------------------------
# STEP 3: Draw line between launch site and selected point
# ------------------------------------------------------

folium.PolyLine(
    locations=[
        [launch_lat, launch_lon],
        [closest_point_lat, closest_point_lon]
    ],
    weight=4,
    color="green"
).add_to(city_map)

# ------------------------------------------------------
# STEP 4: Add distance label on the point
# ------------------------------------------------------

folium.Marker(
    [closest_point_lat, closest_point_lon],
    icon=DivIcon(
        icon_size=(100,20),
        icon_anchor=(0,0),
        html=f'<div style="font-size: 14px; color:#d35400;"><b>{distance_to_point:.2f} KM</b></div>'
    )
).add_to(city_map)


city_map
```
